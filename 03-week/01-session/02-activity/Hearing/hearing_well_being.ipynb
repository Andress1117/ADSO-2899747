```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import numpy as np

# Configurar matplotlib para mostrar texto en espaol
plt.rcParams['font.family'] = 'DejaVu Sans'
plt.rcParams['figure.figsize'] = (12, 8)
plt.style.use('default')

# Leer el archivo CSV
df = pd.read_csv('Hearing well-being Survey Report.csv')

# Diccionario para traducir todas las categoras al espaol
traduccion_categorias = {
    # Grupos de edad
    '18-24': '18-24 aos',
    '25-29': '25-29 aos',
    '30-34': '30-34 aos',
    '35-39': '35-39 aos',
    '40-44': '40-44 aos',
    '45-49': '45-49 aos',
    '50+': '50+ aos',

    # Uso de audfonos
    'Never': 'Nunca',
    'Rarely': 'Raramente',
    'Sometimes': 'A veces',
    'Often': 'Frecuentemente',
    'Daily': 'Diariamente',
    'Always': 'Siempre',
    'Multiple times a day': 'Varias veces al da',
    'Once a day': 'Una vez al da',
    'Few times a week': 'Pocas veces a la semana',
    'Once a week': 'Una vez a la semana',
    'Less than once a week': 'Menos de una vez por semana',

    # Inters en apps
    'Very interested': 'Muy interesado',
    'Interested': 'Interesado',
    'Somewhat interested': 'Algo interesado',
    'Not very interested': 'Poco interesado',
    'Not interested': 'No interesado',
    'Yes': 'S',
    'No': 'No',
    'Maybe': 'Tal vez',

    # Malestar
    'Never': 'Nunca',
    'Rarely': 'Raramente',
    'Sometimes': 'A veces',
    'Often': 'Frecuentemente',
    'Always': 'Siempre',
    'Yes, often': 'S, frecuentemente',
    'Yes, sometimes': 'S, a veces',
    'No': 'No'
}

# Funcin para traducir categoras
def traducir_categorias(serie, diccionario_traduccion):
    return serie.map(diccionario_traduccion).fillna(serie)

# Crear figura con 4 subgrficas
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Anlisis de Encuesta: Bienestar Auditivo', fontsize=16, fontweight='bold')

# Colores personalizados
colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FFEAA7', '#DDA0DD', '#98D8C8']

# GRFICA 1: Distribucin por Grupos de Edad
age_counts = df['Age_group'].value_counts()
# Traducir las categoras de edad
age_counts_es = age_counts.rename(index=traduccion_categorias)
ax1 = axes[0, 0]
bars1 = ax1.bar(range(len(age_counts_es)), age_counts_es.values, color=colors[:len(age_counts_es)])
ax1.set_title('Distribucin de Participantes por Grupo de Edad', fontweight='bold', pad=20)
ax1.set_xlabel('Grupos de Edad')
ax1.set_ylabel('Nmero de Participantes')
ax1.set_xticks(range(len(age_counts_es)))
ax1.set_xticklabels(age_counts_es.index, rotation=45, ha='right')

# Aadir valores en las barras
for i, bar in enumerate(bars1):
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height + 1,
             f'{int(height)}', ha='center', va='bottom', fontweight='bold')

ax1.grid(axis='y', alpha=0.3)

# GRFICA 2: Uso Diario de Audfonos
headphone_counts = df['Daily_Headphone_Use'].value_counts()
# Traducir las categoras de uso de audfonos
headphone_counts_es = headphone_counts.rename(index=traduccion_categorias)
ax2 = axes[0, 1]
bars2 = ax2.bar(range(len(headphone_counts_es)), headphone_counts_es.values, color=colors[:len(headphone_counts_es)])
ax2.set_title('Frecuencia de Uso Diario de Audfonos', fontweight='bold', pad=20)
ax2.set_xlabel('Frecuencia de Uso')
ax2.set_ylabel('Nmero de Participantes')
ax2.set_xticks(range(len(headphone_counts_es)))
ax2.set_xticklabels(headphone_counts_es.index, rotation=45, ha='right')

# Aadir valores en las barras
for i, bar in enumerate(bars2):
    height = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2., height + 1,
             f'{int(height)}', ha='center', va='bottom', fontweight='bold')

ax2.grid(axis='y', alpha=0.3)

# GRFICA 3: Inters en Aplicacin de Audicin
app_interest_counts = df['Interest_in_Hearing_App'].value_counts()
# Traducir las categoras de inters en apps
app_interest_counts_es = app_interest_counts.rename(index=traduccion_categorias)
ax3 = axes[1, 0]
bars3 = ax3.bar(range(len(app_interest_counts_es)), app_interest_counts_es.values, color=colors[:len(app_interest_counts_es)])
ax3.set_title('Inters en Aplicaciones para el Cuidado Auditivo', fontweight='bold', pad=20)
ax3.set_xlabel('Nivel de Inters')
ax3.set_ylabel('Nmero de Participantes')
ax3.set_xticks(range(len(app_interest_counts_es)))
ax3.set_xticklabels(app_interest_counts_es.index, rotation=45, ha='right')

# Aadir valores en las barras
for i, bar in enumerate(bars3):
    height = bar.get_height()
    ax3.text(bar.get_x() + bar.get_width()/2., height + 1,
             f'{int(height)}', ha='center', va='bottom', fontweight='bold')

ax3.grid(axis='y', alpha=0.3)

# GRFICA 4: Malestar en los Odos Despus del Uso
discomfort_counts = df['Ear_Discomfort_After_Use'].value_counts()
# Traducir las categoras de malestar
discomfort_counts_es = discomfort_counts.rename(index=traduccion_categorias)
ax4 = axes[1, 1]
bars4 = ax4.bar(range(len(discomfort_counts_es)), discomfort_counts_es.values, color=colors[:len(discomfort_counts_es)])
ax4.set_title('Malestar en Odos Despus del Uso de Audfonos', fontweight='bold', pad=20)
ax4.set_xlabel('Frecuencia del Malestar')
ax4.set_ylabel('Nmero de Participantes')
ax4.set_xticks(range(len(discomfort_counts_es)))
ax4.set_xticklabels(discomfort_counts_es.index, rotation=45, ha='right')

# Aadir valores en las barras
for i, bar in enumerate(bars4):
    height = bar.get_height()
    ax4.text(bar.get_x() + bar.get_width()/2., height + 1,
             f'{int(height)}', ha='center', va='bottom', fontweight='bold')

ax4.grid(axis='y', alpha=0.3)

# Ajustar espaciado
plt.tight_layout()
plt.show()

# ANLISIS ADICIONAL: Estadsticas resumidas en espaol
print("="*60)
print("RESUMEN EJECUTIVO - ENCUESTA BIENESTAR AUDITIVO")
print("="*60)
print(f"Total de participantes: {len(df)}")
print()

print(" DISTRIBUCIN POR EDAD:")
for age, count in age_counts_es.items():
    percentage = (count/len(df))*100
    print(f"    {age}: {count} personas ({percentage:.1f}%)")
print()

print(" USO DE AUDFONOS:")
for usage, count in headphone_counts_es.items():
    percentage = (count/len(df))*100
    print(f"    {usage}: {count} personas ({percentage:.1f}%)")
print()

print(" INTERS EN APPS DE AUDICIN:")
for interest, count in app_interest_counts_es.items():
    percentage = (count/len(df))*100
    print(f"    {interest}: {count} personas ({percentage:.1f}%)")
print()

print(" MALESTAR AUDITIVO:")
for discomfort, count in discomfort_counts_es.items():
    percentage = (count/len(df))*100
    print(f"    {discomfort}: {count} personas ({percentage:.1f}%)")

# Crear grfica de resumen con los puntos ms importantes
plt.figure(figsize=(14, 8))

# Calcular datos para el resumen (usando las categoras originales para el clculo)
uso_diario = 0
for categoria in df['Daily_Headphone_Use'].value_counts().index:
    if any(palabra in categoria.lower() for palabra in ['daily', 'always', 'multiple']):
        uso_diario += df['Daily_Headphone_Use'].value_counts()[categoria]

interes_apps = 0
for categoria in df['Interest_in_Hearing_App'].value_counts().index:
    if any(palabra in categoria.lower() for palabra in ['yes', 'interested', 'very']):
        interes_apps += df['Interest_in_Hearing_App'].value_counts()[categoria]

malestar = 0
for categoria in df['Ear_Discomfort_After_Use'].value_counts().index:
    if any(palabra in categoria.lower() for palabra in ['yes', 'often', 'sometimes', 'always']):
        malestar += df['Ear_Discomfort_After_Use'].value_counts()[categoria]

jovenes = 0
for categoria in df['Age_group'].value_counts().index:
    if any(edad in categoria for edad in ['18-24', '25-29']):
        jovenes += df['Age_group'].value_counts()[categoria]

# Datos para el resumen en espaol
summary_data = {
    'Usan audfonos\ndiariamente': uso_diario,
    'Interesados en\napps auditivas': interes_apps,
    'Reportan malestar\nauditivo': malestar,
    'Menores de\n30 aos': jovenes
}

categories = list(summary_data.keys())
values = list(summary_data.values())

plt.bar(categories, values, color=['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4'])
plt.title('Puntos Clave: Bienestar Auditivo y Tecnologa', fontsize=16, fontweight='bold', pad=20)
plt.ylabel('Nmero de Participantes')
plt.xticks(rotation=45, ha='right')

# Aadir valores en las barras
for i, v in enumerate(values):
    plt.text(i, v + 5, str(v), ha='center', va='bottom', fontweight='bold', fontsize=12)
    percentage = (v/len(df))*100
    plt.text(i, v/2, f'{percentage:.1f}%', ha='center', va='center', fontweight='bold', color='white', fontsize=10)

plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

print("\n" + "="*60)
print("CONCLUSIONES PRINCIPALES:")
print("="*60)
print("1. La mayora de participantes son jvenes adultos")
print("2. Existe alto uso de dispositivos auditivos diariamente")
print("3. Hay inters significativo en aplicaciones de salud auditiva")
print("4. Un porcentaje considerable reporta malestar auditivo")
print("5. Oportunidad de mercado en el segmento joven para apps auditivas")
```
